```python
df=df.drop("director",axis=1)
```

```python
df
```

```python
df=df.fillna("Inconne")
df
```

```python
df['date_added'] = df['date_added'].str.strip()
df
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce', infer_datetime_format=True)
df
```

```python
print(df['date_added'].isnull().sum())
```

```python
df['year_added'] = df['date_added'].dt.year.astype('Int64')
df['month_added'] = df['date_added'].dt.month.astype('Int64')
df['day_added'] = df['date_added'].dt.day.astype('Int64')
```

```python
# Le graphique montre que la majorité des contenus sur Netflix sont destinés à un public jeune ou familial (TV-MA, TV-14, TV-PG)
df["rating"].value_counts(normalize=True).plot.bar(color='pink')

plt.title("Proportion des contenus par classification d’âge")
plt.xlabel("Rating")
plt.ylabel("Proportion")
plt.xticks(rotation=45)
plt.show()
```

```python
# The graph shows that Netflix has significantly more movies than TV shows in its content library
sns.countplot(data=df, x='type' , palette=["#ff66de",'#ff9999'])
plt.title('Proportion de films vs séries (type)')
plt.show()
```

```python
# Le graphique montre que le nombre de contenus ajoutés a fortement augmenté après 2015, a atteint un pic en 2019, puis a légèrement diminué par la suite
sns.countplot(data=df, x='year_added', color='pink')
plt.title('Nombre de contenus ajoutés chaque année')
plt.xticks(rotation=45)
plt.show()
```

```python
# Le nombre de films et séries augmente fortement après 2015, atteint un pic en 2019, puis diminue légèrement ensuite.
sns.countplot(data=df,x='year_added',hue='type',palette=["#ff66de",'#ff9999'])
plt.title("Évolution du catalogue : films vs séries par année")
plt.xticks(rotation=45)
plt.show()
```

```python
# Le nombre de séries TV augmente fortement jusqu’en 2019–2020, puis diminue légèrement en 2021.
series = df[(df['type']=='TV Show') & (df['year_added']>=2011) & (df['year_added']<=2021)]

x=series["year_added"].value_counts()
sns.lineplot(x=x.index, y=x.values,color='pink')
```

```python
# Les États-Unis dominent largement le nombre de productions, suivis par l’Inde et le Royaume-Uni.
df['country'] = df['country'].str.split(',')
df = df.explode('country')
top10 = df['country'].value_counts().head(10).plot.bar(color='pink')
plt.title("Top 10 des pays par nombre de productions")
plt.show()
```

```python
# Les films sont généralement plus nombreux que les séries TV dans la plupart des pays présentés
top10 = df.head(30)
sns.countplot(data=top10, x='country', hue='type',palette=["#ff66de",'#ff9999'])
plt.title("Nombre de productions par pays et par classification d’âge")
plt.xticks(rotation=45)
plt.show()
```

```python
# Les films internationaux et les drames dominent largement le catalogue en nombre de contenus.
df['listed_in'] = df['listed_in'].str.split(',')
df = df.explode('listed_in')
df['listed_in'].value_counts().head(10).plot.bar(color='pink')
plt.title("Top 10 genres par nombre de contenus")
plt.xlabel("type")
plt.ylabel("Nombre de contenus")
plt.show()
```

```python
# Les films dominent la majorité des genres, tandis que les séries TV sont surtout présentes dans les shows internationaux et les dramas.
df['listed_in'] = df['listed_in'].str.split(',')
df = df.explode('listed_in')

top10 = df['listed_in'].value_counts().head(10).index
df_top10 = df[df['listed_in'].isin(top10)]
sns.countplot(data=df_top10, x='listed_in', hue='type',palette=["#ff66de",'#ff9999'])

plt.title("Top 10 des genres par type")
plt.xlabel("Nombre de contenus")
plt.xticks(rotation=90)
plt.ylabel("type")
plt.show()
```
